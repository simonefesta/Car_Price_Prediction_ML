<h3> Outliers</h3>

Gli outlier sono valori che si discostano significativamente dalla distribuzione normale dei dati. 

Per identificare gli outlier, è comune utilizzare l'intervallo interquartile (IQR), calcolato come segue:
$$\text{IRQ} = \text{Q3} - \text{Q1} $$



Dove **Q1** è il primo quartile (25° percentile) e **Q3** è il terzo quartile (75° percentile)
- Valori inferiori a $\text{Q1} - 1.5 \times \text{IQR}$ sono considerati outlier inferiori.
- Valori superiori a $\text{Q3} + 1.5 \times \text{IQR}$ sono considerati outlier superiori.


```python
# Stampa il numero di righe prima della rimozione degli outliers
print("Numero di righe prima della rimozione degli outliers:", dataset.shape[0])

# Calcola il primo e il terzo quartile per ogni variabile numerica
Q1 = dataset.quantile(0.25, numeric_only=True)
Q3 = dataset.quantile(0.75, numeric_only=True)

# Calcola l'Intervallo Interquartile (IQR) per ogni variabile numerica
IQR = Q3 - Q1

# Allinea i limiti inferiori e superiori con le colonne del DataFrame
Q1_aligned = (Q1 - 1.5 * IQR).reindex(dataset.columns, axis=1)
Q3_aligned = (Q3 + 1.5 * IQR).reindex(dataset.columns, axis=1)

# Identifica le righe che contengono outlier
outliers = dataset[((dataset < Q1_aligned) | (dataset > Q3_aligned)).any(axis=1)]

# Rimuovi gli outliers
dataset_cleaned = dataset[~((dataset < Q1_aligned) | (dataset > Q3_aligned)).any(axis=1)]

# Stampa il numero di righe dopo la rimozione degli outliers
print("Numero di righe dopo la rimozione degli outliers:", dataset_cleaned.shape[0])

# Stampa le righe che sono state rimosse
print("Righe rimosse (outliers):")
print(outliers)
```

<h2> Preparazione modelli di regressione</h2>


<h4> Trasformazione delle variabili categoriche in binarie</h4>

Trasformo le variabili categoriche (come il produttore dell'auto e il tipo di trasmissione) in formato numerico attraverso la tecnica della codifica one-hot. Questo processo crea una colonna separata per ciascuna categoria unica di queste variabili, con valori 0 o 1 che indicano la presenza o l'assenza di quella categoria per ciascuna riga.

```python
# Codifica le variabili categoriche tramite Pandas
dataset_encoded = pd.get_dummies(dataset, columns=['manufacturer', 'model', 'fuelType', 'transmission'])

# Separa le features e il target
X = dataset_encoded.drop('price', axis=1)
y = dataset_encoded['price']
```

<h4> Generazione di training e testing set</h4>

```python
# Suddivisione dei dati in set di addestramento e test, 20% per il testing, il seed è 123456
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=123456)
```

<h4>Normalizzazione</h4>

La standardizzazione trasforma i dati affinché abbiano una media di 0 e una deviazione standard di 1.



```python
# Creazione dello StandardScaler
scaler = StandardScaler()

# Addestramento dello scaler e trasformazione di X_train
X_train_scaled = scaler.fit_transform(X_train) #fit_transform perchè calcolo le statistiche e normalizzo.

# Trasformazione di X_test
X_test_scaled = scaler.transform(X_test) #solo transform perchè non calcolo le statistiche, uso quelle del training.
```

<h2> Addestramento e Predizione</h2>

```python
# Usati per memorizzare e confrontare i risultati.
models=[]
maes=[]
r2s=[]
```

<h3> K-neighbors </h3>

E' un algoritmo utilizzato nel riconoscimento di pattern per la classificazione di oggetti basandosi sulle caratteristiche degli oggetti vicini a quello considerato. In entrambi i casi, l'input è costituito dai k esempi di addestramento più vicini nello spazio delle funzionalità. L'output dipende dall'utilizzo di k-NN per la classificazione o la regressione:

```python
# Creazione e ricerca dei migliori parametri per K-Nearest Neighbors
param_grid = {'n_neighbors': np.arange(1, 21)}

random_search = RandomizedSearchCV(
    KNeighborsRegressor(),                    # Il modello di regressione KNN da ottimizzare
    param_distributions=param_grid,           # La griglia dei parametri da esplorare
    n_iter=10,                                # Numero di combinazioni di parametri da testare (10)
    cv=5,                                     # Numero di fold per k-fold cross validation (5)
    scoring='neg_mean_absolute_error',        # Metodologia per valutare le prestazioni dei modelli (MAE negativo)
    n_jobs=-1,                                # Numero di CPU da utilizzare (qui, tutti disponibili)
    random_state=123456                       # Seme per garantire la riproducibilità dei risultati
)
random_search.fit(X_train_scaled, y_train)

# Ottieni e stampa i migliori parametri
best_k = random_search.best_params_['n_neighbors']
print("Best hyperparameters:", random_search.best_params_)

# Addestra il modello con i migliori parametri
knn_model = KNeighborsRegressor(n_neighbors=best_k)
knn_model.fit(X_train_scaled, y_train)

# Previsioni e valutazione
y_predict = knn_model.predict(X_test_scaled)
train_y_pred = knn_model.predict(X_train_scaled)
knn_mae = mean_absolute_error(y_test, y_predict)
knn_r2 = r2_score(y_test, y_predict)
maet = mean_absolute_error(y_train, train_y_pred)
r2t = r2_score(y_train, train_y_pred)

print(f'mae_testing: {knn_mae:.2f}')
print(f'r2_testing: {knn_r2:.2f}')
print(f'mae_training: {maet:.2f}')
print(f'r2_training: {r2t:.2f}')

# Memorizza i risultati
models.append('K-Nearest Neighbors')
maes.append(knn_mae)
r2s.append(knn_r2)

# Visualizzazione dei risultati
plt.figure(figsize=(10, 6))
sns.regplot(x=y_test, y=y_predict, scatter_kws={'s':50, 'color':'blue'}, line_kws={'color':'red', 'linewidth':2})
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Comparison of Predicted Values and Actual Values', fontsize=16)
plt.grid(True)
plt.show()
```

<h3> Regressione Lineare</h3>

<h4>Descrizione della Regressione Lineare</h4> 

La **regressione lineare** è una tecnica statistica che modella la relazione tra una variabile dipendente e una o più variabili indipendenti attraverso una retta. Essa stima i parametri del modello per minimizzare la differenza tra i valori osservati e quelli previsti dal modello.

- **Predizione**: Utilizza i parametri stimati per fare previsioni sulla variabile dipendente.
- **Calcolo degli Errori**: Misura l'accuratezza del modello con metriche come l'**Errore Assoluto Medio (MAE)**, che calcola la media delle differenze assolute tra i valori osservati e previsti, e l'**Errore Quadratico Medio (MSE)**, che misura la media dei quadrati di queste differenze. Questi errori aiutano a valutare quanto bene il modello si adatta ai dati.

```python
# Crea il modello di regressione lineare
model = LinearRegression()

# Allena il modello sui dati di addestramento (non normalizzati)
model.fit(X_train, y_train)

# Effettua previsioni sui dati di test (non normalizzati)
y_pred = model.predict(X_test)

# Valuta le prestazioni del modello
mae = mean_absolute_error(y_test, y_pred)
#mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Stampa i risultati
print(f"Mean Absolute Error: {mae}")
#print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")


#Salvataggio per confronto
models.append('LinearRegression')
maes.append(mae)
r2s.append(r2)

# Visualizzazione dei risultati
# Regressione Lineare su Valori Reali vs Predetti
plt.figure(figsize=(10, 6))
sns.regplot(x=y_test, y=y_pred, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Regression Line: Actual vs Predicted Price')
plt.show()

```

Il modello di regressione lineare senza normalizzazione ha fornito i seguenti risultati:

- **Mean Absolute Error (MAE)**: 2034.66
  - Indica che, in media, le previsioni differiscono di circa 2034 unità dai valori effettivi.

- **R² Score**: 0.87
  - Indica che circa l'87% della variabilità nei dati è spiegata dal modello, suggerendo una buona capacità predittiva.

Nel complesso, il modello ha ottenuto buone prestazioni, spiegando una gran parte della variabilità nei dati.

<h3> Ridge, L2 Regularization</h3>

La Regressione Ridge (L2 Regularization) è una tecnica di regressione lineare che include un termine di penalizzazione sui coefficienti del modello. Questa penalizzazione aiuta a ridurre l'overfitting e a migliorare la generalizzazione, soprattutto quando ci sono molte caratteristiche o dati altamente correlati. L2 Regularization aggiunge la somma dei quadrati dei coefficienti al termine di costo, penalizzando i coefficienti grandi e mantenendoli più piccoli.

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=1.0)  # alpha è il parametro di regolarizzazione

# Allena il modello sui dati di addestramento
ridge_model.fit(X_train_scaled, y_train)

# Effettua previsioni sui dati di test
y_pred = ridge_model.predict(X_test_scaled)
# Effettua previsioni sui dati di addestramento
train_y_pred = ridge_model.predict(X_train_scaled)

# Valuta le prestazioni del modello sul set di test
ridge_mae = mean_absolute_error(y_test, y_pred)
ridge_r2 = r2_score(y_test, y_pred)
print(f'mae_testing: {ridge_mae:.2f}')
print(f'r2_testing: {ridge_r2:.2f}')

# Valuta le prestazioni del modello sul set di addestramento
maet = mean_absolute_error(y_train, train_y_pred)
r2t = r2_score(y_train, train_y_pred)
print(f'mae_training: {maet:.2f}')
print(f'r2_training: {r2t:.2f}')

# Memorizza i risultati
models.append('Ridge Regression')
maes.append(ridge_mae)
r2s.append(ridge_r2)

# Visualizzazione dei risultati
plt.figure(figsize=(10, 6))  # Dimensione della figura

# Creazione del grafico di regressione
sns.regplot(x=y_test, y=y_pred, scatter_kws={'s':50, 'color':'blue'}, line_kws={'color':'red', 'linewidth':2})

# Miglioramento dell'aspetto del grafico
plt.xlabel('Actual Price', fontsize=14)
plt.ylabel('Predicted Price', fontsize=14)
plt.title('Comparison of Predicted Values and Actual Values', fontsize=16)
plt.grid(True)  # Aggiungi una griglia

plt.show()
```
